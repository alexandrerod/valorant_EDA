```python
all_teams = pd.read_csv('teams.csv')
all_teams
```

```python
all_teams.info()
```

```python
colunas = ['maps_won%','atk_won%','def_won%','pistol_won%']
all_teams[colunas] = all_teams[colunas].applymap(lambda x: float(x.replace('%', '')))
```

```python
all_teams.info()
```

Como a porcentagem de algumas features estavam com o tipo object definido, precisei utilizar uma função para retirar o sinal de % das linhas e depois converter de object para float para elaborar as análises gráficas.

```python
teams_by_country = all_teams.groupby('country')['team'].size().reset_index()
teams_by_country.sort_values('team', ascending=False).head(10)
```

```python
plt.figure(figsize=(20,10))

sns.barplot(x=teams_by_country['country'][0:10], y= teams_by_country['team'],palette='rocket')
plt.ylabel('Times', fontsize=20)
plt.xlabel('Países', fontsize=20)
plt.title('Distribuição de times por país', fontsize=25)
```

Assim como o canadá tem o maior número de jogadores, eles também tem o maior número te times de valorant, seguido pelo Brasil como a segunda maior região com times

```python
team_def = all_teams.groupby('team')['def_won%'].sum().reset_index()
plt.figure(figsize=(30,10))
best_def = team_def.sort_values('def_won%', ascending=False)
best_def

sns.barplot(x=best_def['team'][0:10], y=best_def['def_won%'], palette='rocket')
plt.ylabel('Vitórias como defensor %', fontsize=20)
plt.xlabel('Times', fontsize=20)
plt.title('Porcentagem de vitórias na defesa de cada time', fontsize=25)
```

O time Kru Fem é o time com melhor desempenho jogando na defesa

```python
team_atk = all_teams.groupby('team')['atk_won%'].sum().reset_index()
plt.figure(figsize=(30,10))
best_atk = team_atk.sort_values('atk_won%', ascending=False)
best_atk

sns.barplot(x=best_atk['team'][0:10], y=best_atk['atk_won%'], palette='rocket')
plt.ylabel('Vitórias como atacante %', fontsize=20)
plt.xlabel('Times', fontsize=20)
plt.title('Porcentagem de vitórias no ataque de cada time', fontsize=25)
```

O time Kru Fem também é o time com melhor desempenho atacando

```python
maps_won = all_teams.groupby('team')['maps_won%'].sum().reset_index()
plt.figure(figsize=(30,10))
team_winner =maps_won.sort_values('maps_won%', ascending=False)
best_atk

sns.barplot(x=team_winner['team'][0:10], y=team_winner['maps_won%'], palette='rocket')
plt.ylabel('Vitórias por mapa %', fontsize=20)
plt.xlabel('Times', fontsize=20)
plt.title('Porcentagem de vitórias por mapa de cada time', fontsize=25)
```

MLC Esports, KRU Fem e DRX VS são os times com maior porcentagem de vitória por mapa que jogaram.

```python
maps_played = all_teams.groupby('team')['maps_played'].sum().reset_index()
plt.figure(figsize=(30,10))
maps =maps_played.sort_values('maps_played', ascending=False)

sns.barplot(x=maps['team'][0:10], y=maps['maps_played'], palette='rocket')
plt.ylabel('Partidas jogadas', fontsize=20)
plt.xlabel('Times', fontsize=20)
plt.title('Quantidade de partidas jogadas por time', fontsize=25)
```

Podemos ver que o time com maior quantidade de partidas jgoadas é o  Knights Academy, podemos supor que é um time criado a mais tempo que os outros, ou que joga mais campeonatos juntos
